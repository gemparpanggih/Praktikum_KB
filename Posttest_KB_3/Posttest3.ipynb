```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("premiere-league.csv")
```

```python
df.dtypes
```

```python
df.head()
```

## Grafik Batang yang menampilkan Pebandingan Hasil Pertadingan Liga Inggris 

### Dari grafik batang di bawah bisa dilihat bahwa banyak team memenangankan pertandingan dalam kandang mereka sendiri ketimbang memenangkan pertandingan di luar kandang mereka. 

### Keterangan Grafik
#### H = Kemenangan Tim Tuan Rumah
#### A = Kemenangan Tim Luar Kandang
#### D = Hasil seri antara Tim Tuan Rumah dan Tim Luar Kandang

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x= 'result', palette='flare')
plt.title("Hasil Pertandingan EPL Season 2006-2018")
plt.xlabel('Hasil Pertandingan')
plt.ylabel('Kuantitas')
plt.show()
```

## Grafik Pie yang menampilkan Jumlah Pertandingan Setiap Season

### Dari grafik pie di bawah bisa dilihat bahwa jumlah pertandingan setiap season dari Season 2006-2007 sampai Season 2017-2018 adalah konsisten, dengan jumlah presentase tiap season adalah 8.3%

```python
data_pie = df["season"].value_counts().rename_axis("season").reset_index(name="id_count")
```

```python
plt.figure(figsize=(8,8))
plt.pie(data_pie["id_count"],labels=data_pie["season"],startangle=120,autopct='%.1f%%')
plt.xlabel('Season EPL')
plt.ylabel('Kuantitas')
plt.show()
```

## Grafik Heatmap yang menampilkan Korelasi perbandingan perbedaan antara Home Goal dan Awal Goal

### Dari hasil di bawah bisa dilihat ketika pertandingan kandang tandang, banyak team membuat goal di kandang mereka sendiri dari pada di kandang team lain

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, cmap='Blues')
ax.set_title('Korelasi Jumlah antara Goal Dalam Kandang dan Goal Luar Kandang')
```

## Grafik Scatter Plot yang menampilkan apakah terdapat hasil pertandingan dalam setiap season EPL

```python
sns.scatterplot(data=df, x=df['season'], y=df['result'])
```
